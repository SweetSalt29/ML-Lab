```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df.head()
```

```python
df.isnull().sum()
```

```python
df.loc[df.index % 2 == 0, 'sepal length (cm)'] = np.nan
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
display(df['sepal length (cm)'])
```

```python
mean_sepal_length = df['sepal length (cm)'].mean()
df['sepal length (cm)'].fillna(mean_sepal_length, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['sepal length (cm)']
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 3)
sns.boxplot(y=df['petal length (cm)'])
plt.title('Box Plot - Petal Length (cm)')

plt.subplot(2, 2, 4)
sns.boxplot(y=df['petal width (cm)'])
plt.title('Box Plot - Petal Width (cm)')

plt.tight_layout()
plt.show()
```
