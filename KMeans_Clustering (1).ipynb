```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data={
    'Cust_ID': list(range(1,21)),
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female', 'Female', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Female', 'Male',
               'Male', 'Female', 'Male', 'Male', 'Female'],
    'Age': [19,21,20,23,31,22,35,23,64,30,67,35,58,24,37,22,35,20,52,35],
    'Annual Income (K$)': [15,15,16,16,17,17,18,18,19,19,19,19,20,20,20,21,21,23,23,23],
    'Spending Score(1-100)': [39,81,6,77,40,76,6,94,3,72,14,99,15,77,13,79,35,66,29,98]
}
```

```python
df=pd.DataFrame(data)
```

```python
#Select Features
X=df[['Annual Income (K$)', 'Spending Score(1-100)']]
```

```python
#Elbow Method to find optimal K
wcss=[]
for i in range(1,11): #Try cluster sizes 1 to 10
    kmeans= KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-means (assume 3 clusters)
kmeans=KMeans(n_clusters=3, init='k-means++', random_state=42)
df['Cluster']=kmeans.fit_predict(X)
```

```python
#Visualize
plt.figure(figsize=(8, 6))
plt.scatter(X['Annual Income (K$)'], X['Spending Score(1-100)'], c=df['Cluster'], cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids', alpha=0.7, edgecolors='k')
plt.title('Customer Segments (K-Means Clustering)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
